```
# importando modulos necesarios
import matplotlib.pyplot as plt 
import numpy as np # importando numpy
import pandas as pd # importando pandas
from scipy import stats
from sympy import *

np.random.seed(111984000) # para poder replicar el random

def calcularError(resultado):
  x = symbols('x')
  f = 1-x**2
  valor = integrate(f,(x,0,1)).evalf()
  error = error = (valor-resultado)*100
  print('Error de:', error.evalf(), '%')

def calculoDeLaIntegral(N=100000):
    plt.figure(figsize=(10,10))  
    x,y = np.random.uniform(0, 1, size=(2, N))
    print('x')
    plt.plot(x, y, 'r.')
    plt.plot(x , 1-x**2, 'b.')
    print("La integral es: ",(sum(y<(1-x**2)))/N)
    a = sum(y<(1-x**2))/N
    plt.axis('square')
    return a

resultado = calculoDeLaIntegral()
calcularError(resultado)
```
