```
# importando modulos necesarios
import matplotlib.pyplot as plt 
import numpy as np # importando numpy
import pandas as pd # importando pandas
from scipy import stats
from sympy import *

np.random.seed(111984000) # para poder replicar el random

def calculoDeLaIntegral(N=10000):
    plt.figure(figsize=(10,10))  
    x, y = np.random.uniform(0, 1, size=(2, N)) 
    plt.plot(x, y, 'r.')
    plt.plot(x , 1-x**2, 'b.')
    print("La integral es: ",(sum(x<(1-x**2)))/N)
    a = sum(x<(1-x**2))/N
    plt.axis('square')
    return a
```

```
x = symbols("x")
f = 1-x**2
valor1 = integrate(f,(x,0,1)).evalf(2)
error = (valor1-calculoDeLaIntegral())*100
print('Error de:', error.evalf(3), '%')
```

```
calculoDeLaIntegral(N=100000)
```
