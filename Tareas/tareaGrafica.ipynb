```python
import pandas as pd
fifa = pd.read_csv ('/content/fifa_datos.csv')
fifa.head(5)
```

```python
fifa2 = fifa['Weight'].interpolate()
M = fifa2.to_numpy()
pesos = []
for i in range(0,len(M)):
  if M[i] != 0:
    x = M[i].split('l')
    pesos.append(int(x[0]))

peso1 = []
peso2 = []
peso3 = []
peso4 = []

for j in range(0, len(pesos)):
  if pesos[j] < 125:
    peso1.append(pesos[j])
  if pesos[j] > 125 and pesos[j] < 150:
    peso2.append(pesos[j])
  if pesos[j] > 150 and pesos[j] < 175:
    peso3.append(pesos[j])
  if pesos[j] > 175:
    peso4.append(pesos[j])
  



```

```python
import matplotlib.pyplot as plt

plt.figure()

etiquetas = ['Peso 1', 'Peso 2', 'Peso 3', 'Peso 4']
colores = ['#abcdef', '#aabbcc', '#babbac', '#babbca']
plt.pie([len(peso1), len(peso2),len(peso3),len(peso4),], labels=etiquetas, colors=colores, autopct='%.2f %%')
plt.title('Pesos')
plt.show()
```
